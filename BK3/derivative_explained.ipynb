# Understanding Derivatives: y = x²

A **derivative** tells us the *rate of change* of a function — in plain English, **how fast y is changing as x changes**.

Think of it like speed: if y is your position, the derivative is your velocity.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 300)
y = x ** 3

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2.5, label=r'$y = x^2$')
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.grid(True, alpha=0.3)
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.title('The Function: y = x²', fontsize=16)
plt.legend(fontsize=14)
plt.tight_layout()
plt.show()
```

## What is the Derivative?

The derivative of $y = x^2$ is:

$$\frac{dy}{dx} = 2x$$

### What does this mean?

| x value | Derivative (2x) | Meaning |
|---------|-----------------|----------|
| -3 | -6 | y is **decreasing steeply** |
| -1 | -2 | y is **decreasing gently** |
| 0 | 0 | y is **flat** (minimum point!) |
| 1 | 2 | y is **increasing gently** |
| 3 | 6 | y is **increasing steeply** |

- **Negative derivative** → the curve is going **downhill** (left side of the parabola)
- **Zero derivative** → the curve is **flat** (the bottom of the U)
- **Positive derivative** → the curve is going **uphill** (right side of the parabola)

```python
dy_dx = 2 * x

fig, axes = plt.subplots(2, 1, figsize=(10, 10), sharex=True)

# Top plot: the original function
axes[0].plot(x, y, 'b-', linewidth=2.5, label=r'$y = x^2$')
axes[0].axhline(0, color='gray', linewidth=0.5)
axes[0].axvline(0, color='gray', linewidth=0.5)
axes[0].set_ylabel('y', fontsize=14)
axes[0].set_title('Original Function', fontsize=16)
axes[0].legend(fontsize=14)
axes[0].grid(True, alpha=0.3)

# Bottom plot: the derivative
axes[1].plot(x, dy_dx, 'r-', linewidth=2.5, label=r"$\frac{dy}{dx} = 2x$")
axes[1].axhline(0, color='gray', linewidth=0.5)
axes[1].axvline(0, color='gray', linewidth=0.5)
axes[1].fill_between(x, dy_dx, 0, where=(dy_dx > 0), alpha=0.15, color='green', label='Increasing')
axes[1].fill_between(x, dy_dx, 0, where=(dy_dx < 0), alpha=0.15, color='red', label='Decreasing')
axes[1].set_xlabel('x', fontsize=14)
axes[1].set_ylabel('dy/dx', fontsize=14)
axes[1].set_title('Derivative (slope at each point)', fontsize=16)
axes[1].legend(fontsize=14)
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## Visualizing the Slope as a Tangent Line

The derivative at a specific point gives the **slope of the tangent line** at that point. Let's see tangent lines at a few key locations.

```python
plt.figure(figsize=(10, 7))
plt.plot(x, y, 'b-', linewidth=2.5, label=r'$y = x^2$')

points = [-3, -1, 0, 1, 3]
colors = ['#e74c3c', '#e67e22', '#2ecc71', '#3498db', '#9b59b6']

for xi, c in zip(points, colors):
    yi = xi ** 2
    slope = 2 * xi
    x_tan = np.linspace(xi - 1.5, xi + 1.5, 50)
    y_tan = slope * (x_tan - xi) + yi

    plt.plot(x_tan, y_tan, '--', color=c, linewidth=2,
             label=f'x={xi}, slope={slope}')
    plt.plot(xi, yi, 'o', color=c, markersize=10, zorder=5)

plt.xlim(-5, 5)
plt.ylim(-5, 25)
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.grid(True, alpha=0.3)
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.title('Tangent Lines Show the Derivative at Each Point', fontsize=16)
plt.legend(fontsize=12, loc='upper left')
plt.tight_layout()
plt.show()
```

## Key Takeaways

1. The **derivative** of $y = x^2$ is $\frac{dy}{dx} = 2x$
2. It tells you the **slope** (steepness) of the curve at every point
3. Where the derivative is **zero**, the function has a **minimum** (or maximum)
4. The derivative is itself a function — it maps each x to the slope at that x
5. Geometrically, the derivative equals the slope of the **tangent line** touching the curve
